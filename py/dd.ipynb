```python
def is_valid_triangle(side1, side2, side3):
    return side1 + side2 < side3 or \
        side1 + side3 < side2 or \
        side3 + side2 < side1
```

```python
def is_prime(num):
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

n = int(input())

print(is_prime(n))
```

```python
def is_palindrome(text):
    n = len(text)
    text.lower()
    text.replace('')
    
    for i in range(n // 2):
        if text[i] != text[n - i - 1]:
            return False
    return True
```

```python
def convert_to_python_case(text):
    words = text.split('_')
    text = ''.join(word.ca for word in words)
    return text
```

```python
def draw_triangle(base, height):
    for i in range(height):
        print(' ' *  (base // 2 - i)+ '*' * ((i * 2) + 1))

draw_triangle(15, 8)
```

```python
left = int(input())
right = int(input())
flag = True
if left > right:
    flag = False

num = input()
while num:
    num = int(num)
    if left <= num <= right:
        flag = flag and True
    else: 
        flag = False
    num = input()

print(flag)
```
